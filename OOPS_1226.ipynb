

# **Hierarchical** **Inheritance**

Hierarchical inheritance is a type of inheritance where **multiple child classes inherit from a single parent class**.

It means one base (parent) class is shared by more than one derived (child) class, and each child can have its own additional features while still using the common properties and methods of the parent.

**Example**

It’s like one parent having multiple children — all children inherit some common traits, but each child can also have unique characteristics.


```python
# Parent class
class Animal:
    def eat(self):
        print("Animal eats food")

# Child class 1
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Child class 2
class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Creating objects
dog = Dog()
cat = Cat()

# Using inherited method
dog.eat()   # from Animal
cat.eat()   # from Animal

# Using their own methods
dog.bark()
cat.meow()
```

# Hybrid inheritance

combo of more than one inheritance, single+mullti like that

```python
# Base class
class Animal:
    def eat(self):
        print("Animal eats food")

# Derived from Animal (hierarchical part)
class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Another base class
class Pet:
    def play(self):
        print("Pet plays")

# Hybrid class (multiple inheritance + hierarchical)
class Puppy(Dog, Pet):
    def weep(self):
        print("Puppy weeps")

# Object
p = Puppy()

# Accessing methods
p.eat()    # from Animal
p.bark()   # from Dog
p.play()   # from Pet
p.weep()   # own method
```

we assumed Python would treat both methods as separate versions (like overloading). It didn’t.

When you defined the method the second time, Python **replaced the first one completely**. It doesn’t store multiple versions based on parameters—only the **last definition exists**.

So your class ended up with just one method that expects **three arguments**.

Then you called it with only **two arguments**, which is why Python threw an error saying one argument is missing.

That’s the whole issue:

* No overloading in Python
* Last method wins
* Argument mismatch causes the error

```python
# handling method overriding init method
# 1st way
class Parent:
    def __init__(self, name):
        self.name = name

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)   # call parent init
        self.age = age
```
